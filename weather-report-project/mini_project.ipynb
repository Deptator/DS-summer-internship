```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

### Phase 1- Load the data

```python
url = "https://raw.githubusercontent.com/rajendra0968jangid/Ds-arya/main/temperature_data.json"
df = pd.read_json(url)

print(df)

print(df.head())
print(df.tail())

print("\nDescription:")
print(df.describe())
print("\nInformation:")
df.info()
```

### Phase-2: Data Inspection

```python
# shows the missing values in each column
print(df.isnull().sum())

df.describe()
print("\n Shape: ",df.shape)
df.columns
```

### Phase-3: Data cleaning

```python
# 
print(df)

df.drop(2, inplace=True, errors="ignore")

# calculating the humidity mean 
humidity_mean = df["humidity_pct"].mean()
print("\nHumidity mean:",humidity_mean)

# filling the humidity NaN
df["humidity_pct"] = df["humidity_pct"].fillna(humidity_mean)
df.isnull().sum()

# output 
# day              0
# temperature_c    0
# humidity_pct     0
# condition        0
# dtype: int64


# adding Farenheit column
df["temperature_f"] = (df["temperature_c"] * 1.8) + 32
print(df)

print(df.info())
print(df.head())
```

### Phase-4: Visualization


```python
plt.style.use("dark_background")

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Line Chart
ax[0].plot(df["day"], df["temperature_c"],
           marker='o', label="Temperature (C)")
ax[0].set_title("Temperature Trend")
ax[0].set_xlabel("Day")
ax[0].set_ylabel("Temperature (C)")
ax[0].grid(True)
ax[0].legend()

# Pie Chart
count = df["condition"].value_counts()

ax[1].pie(count,labels=count.index,autopct='%1.1f%%')
ax[1].set_title("Weather Conditions")

plt.tight_layout()

# Save Image
plt.savefig("weather_report.jpg")

plt.show()

```

## FINAL REPORT

```python
# Final Report
print("\nAverage Temperature:", df["temperature_c"].mean())
print("Highest Temperature:", df["temperature_c"].max())
print("Lowest Temperature:", df["temperature_c"].min())
print("Average Humidity:", df["humidity_pct"].mean())
print("Most Common Condition:", df["condition"].mode()[0])
```

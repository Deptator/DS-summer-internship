```python
import numpy as np 
```

### creating array from list

```python
arr_1d = np.array([1,2,3,4])
print("1D array: ", arr_1d)

arr_2d = np.array([[1,2,3],[4,5,6]]) # try to use same type of data in order to perform execution of data faster.
print("2D array: ",arr_2d)
```

### list vs numpy array

```python
py_list = [1, 2, 3]
print("Python list multiplication ", py_list * 2 )

np_array = np.array([1, 2, 3]) # element wise multiplication
print("Python array multiplication ", np_array * 2)

import time
start = time.time()
py_list = [i*2 for i in range (1000000)]
print("\n List operation time: ", time.time()-start)

start = time.time()
np_array = np.arange(100000) * 2
print("\n Numpy operation time: ", time.time()-start)
```

### creating array from scratch

```python
zeros = np.zeros((3,4)) # creates a matrice of element 0 of required rows and column
print("Zeros array: \n", zeros)

ones = np.ones((2,3)) # creates a matrice of element 1 of required rows and column
print("One array: \n", ones)

full = np.full((2, 2), 7) # creates a matrice of constant element of required row and column
print("Full array: \n", full)

random = np.random.random((2, 3))
print("Random array: \n", random)

sequence = np.arange(0, 11, 2)
print("Sequence array: \n", sequence)

```

### Vector, Matrix and Tensor

```python
vector = np.array([1, 2, 3])
print("Vector: ", vector)

matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])
print("Matrix: ", matrix)

tensor = np.array([[[1,2], [3,4]],
                   [[5,6],[7,8]]]) # if more than 2D array then we call it as tensor or simply call it multi dimensional array
print("Tensor: ",tensor)
```

### Array Properties

```python
arr = np.array([[1, 2, 3],
               [4, 5, "6.8"]])

print("Shape:", arr.shape)
print("Dimension: ", arr.ndim)
print("Size: ", arr.size)
print("DataType: ", arr.dtype)
```

### Array Reshaping

```python
arr = np.arange(12)
print("Original Array: ", arr)

reshaped = arr.reshape((3,4))
print("Reshaped Array: \n", reshaped)

flattened = reshaped.flatten()
print("Flattened Array: ", flattened)

# ravel (returns view, instead of copy)
raveled = reshaped.ravel()
print("\n Raveled Array: ", raveled)

# Transpose 
transpose = reshaped.T
print("\n Transpose Array: \n", transpose)
```

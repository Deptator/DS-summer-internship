## Data Preprocessing


```python
import pandas as pd

# importing sklearn ways

# import sklearn
# import sklearn as sk 
# from sklearn import labelencoder
# from sklearn.preprocessing import *
from sklearn.preprocessing import LabelEncoder
```

```python
data = {
    "Name" : ['Ram', 'Kamal', 'Ajay'],
    'Age': [25,None, 32],
    'Salary':[5000, 6000, None]
}
df = pd.DataFrame(data)
print(f"Original Data Frame:\n{df}")
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
print(df)
```

### Encoding categorical values


```python
data2 = {
    "Name" : ['Ram', 'Kamal', 'Ajay', 'Neetu', 'Kavi', 'Sapu'],
    'Age': [25, None, 32, 20, 21, 22],
    'Salary':[5000, 6000, None, 7000, 8000, 9000],
    'Gender' : ['Male', 'Male', 'Male', None, 'Female', 'Female']
}
df2 = pd.DataFrame(data2)
print(f"Original Data Frame: \n{df2}")
le = LabelEncoder()
df2_label = df2.copy()
df2_label['Gender'] = df2_label['Gender'].fillna('unknown')
df2_label['Gender_Encoder'] = le.fit_transform(df2_label['Gender'])

print("\nAfter Encoding:")
print(df2_label)
```

```python
# combination of handle missing value | label encoding | onehot encoding

import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

data = {
    "age":[10,20,np.nan,26,30],
    "gender":["male","female","others","male",None],
    "name":["dheeraj","kavi","sapu","yash","hello"]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
```

```python
# age column -> fill with mean

df['age'] = df['age'].fillna(df['age'].mean())

# gender column -> fill with unknown

df['gender'] = df['gender'].fillna('unknown')

print("\nAfter handling missing values:")
print(df)
```

```python
le = LabelEncoder()

df['gender_label'] = le.fit_transform(df['gender'])

print("\nAfter Label Encoding:")
print(df)
```

```python
ohe = OneHotEncoder(sparse_output=False)

encoded = ohe.fit_transform(df[['gender']])

encoded_df = pd.DataFrame(encoded,columns=ohe.get_feature_names_out(['gender']))

print("\nOne Hot Encoded:")
print(encoded_df)
```

```python
final_df = pd.concat([df, encoded_df], axis=1)

print("\nFinal DataFrame:")
print(final_df)
```

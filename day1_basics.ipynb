## 0. importing matplot 

```python
import matplotlib.pyplot as plt
```

## Creating the first graph

```python
x = [2010, 2015, 2020, 2025] # x cord
y = [100, 200, 250, 300] # y cord
plt.plot(x,y) # Line graph

plt.xlabel("year") # labeling the x cord
plt.ylabel("sales") # labeling the y cord
plt.title("Sales report") # graph title
plt.show() # showing the graph
```

### modifying the height and width of the graph

```python
x = [2010, 2015, 2020, 2025] # x cord
y = [100, 200, 250, 300] # y cord

plt.figure(figsize=(6,2)) # (width, height)
plt.plot(x,y)
plt.show()
```

### Modifying the color of the line and marker styles


```python
plt.plot(x,y,color='y', marker='o', linestyle='dotted', linewidth=0.8, markersize=5)
```

## 2. Multiple lines & Legends

```python
# multi line chart

x = [2010, 2015, 2020, 2025]
y1 = [100, 200, 260, 290]
y2 = [150, 185, 195, 300]

plt.plot(x, y1, label="jeans",marker = '1')
plt.plot(x, y2, label="shirt", marker='o', linestyle ='dotted')
plt.legend()
plt.show()
```

```python

x = [2010, 2015, 2020, 2025]
y1 = [100, 200, 260, 210]
y2 = [150, 185, 195, 300]

plt.plot(x, y1, label="Product A", color='y', marker='o')

plt.plot(x, y2, label="Product B", color= 'c', marker='1')
plt.style.use('dark_background')
plt.grid("True")
plt.title("Product Growth")
plt.xlabel("Year")
plt.ylabel("Revenue")

plt.legend()

plt.show()
```

## Bar chart

```python
x = [2010, 2015, 2020, 2025]
y1 = [100, 200, 260, 210]

plt.style.use('dark_background')
plt.figure(figsize=(10,5))
plt.bar(x,y1, color = 'b')
plt.show()
```

```python
x = ['Group A', 'Group B', 'Group C', 'Group D']
y1 = [10, 20, 20, 40]
y2 = [20, 30, 40, 45]

plt.title("")
plt.bar(x,y1)
plt.bar(x,y2)
```
